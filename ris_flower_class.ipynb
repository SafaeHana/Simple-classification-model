```python
#librairies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as ps
%matplotlib inline
```

```python
columns=['Sepal length','Sepal width','Petal lenght','Petal width','class_labels']
#loading the data
df = pd.read_csv('C:/Users/safae/Documents/iris.data',names=columns)
df.head()
```

```python
df.describe()
```

```python
#Visualize the whole dataset
ps.pairplot(df,hue='class_labels')
```

```python
#Separate features and target
data =df.values
X= data[:,0:4]
Y= data[:,4]
print(Y)
```

```python
#Splitting the data into training and testing dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, Y, test_size=0.2)
print(X_train)
```

```python
# Support vector ML algorithm
from sklearn.svm import SVC
model_svc = SVC()
model_svc.fit(X_train, y_train)
```

```python
prediction1 = model_svc.predict(X_test)
# calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction1)*100)
```

```python
#Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
#A Decision Tree Classifier is a computer program that plays this game using a set of rules to make decisions. In this case, we want to use a Decision Tree Classifier to guess the answer to a question based on some data we have about the question. 
model_DTC=DecisionTreeClassifier()
model_DTC.fit(X_train, y_train)
```
